```python
import pandas as pd
df = pd.read_csv("metrics_log.csv")
df.head()
```

```python
test_df = df[df["Split"] == "Test"].copy()
# test_df = test_df[pd.to_datetime(test_df["WindowEnd"]) == pd.to_datetime("2020-01-04")]
test_df.head()
```

```python
agg_metrics = test_df.groupby("Tag")[["IC", "IR", "CR", "Prec@K"]].mean().reset_index()
agg_metrics.head(15)
```

```python
metrics_to_sort = ["IC", "IR", "CR", "Prec@K"]

sorted_results = {}

for metric in metrics_to_sort:
    sorted_df = agg_metrics.sort_values(by=metric, ascending=False).reset_index(drop=True)
    sorted_results[metric] = sorted_df
    print(f"\n🔹 Top Models by {metric}")
    display(sorted_df)
```
